```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
```

## Plot english A's and french A's against the total amount of Characters

```python
french = np.loadtxt("Dataset french.txt") 
english = np.loadtxt("Dataset english.txt") 

fr_charac =french[:,0]
fr_A = french[:,1]
en_charac = english[:,0]
en_A = english[:,1]

plt.plot(fr_charac, fr_A, 'r.', label='french')
plt.plot(en_charac, en_A, 'b.', label='english')
x = np.arange(0,80000,10)
print(x)
y = []
for i in x: 
    print
    y.append(1/(1+exp(-(0.06517381*i+18.80914014))))
    

plt.plot(x, y)
plt.legend(loc ='lower right')
plt.xlabel('Characters')
plt.ylabel('A\'s')
plt.show()
```
